```python
import csv    
```

```python
# Preamble
data = []
headers = []
with open('PeopleTrainngDate.csv', 'r') as file:
    csv_reader = csv.reader(file)
    headers = next(csv_reader)
    colLen = [len(h) for h in headers]
    
    for row in csv_reader:
        if len(row) == 0:
            continue
        data.append(row)
        for i, e in enumerate(row):
            if len(e) > colLen[i]:
                colLen[i] = len(e)
```

```python
# Excercise 1
print("  ".join(f"{item:<{colLen[i]}}" for i, item in enumerate(headers)))
print("-"*(sum(colLen)+(2*(len(colLen)-1))))
for row in data:
    print("  ".join(f"{item:<{colLen[i]}}" for i, item in enumerate(row)))
```

```python
# Exercise 2
def sortFunc(x):
    d = x[5].split('/')
    return f'{d[2]}{d[1]}{d[0]}'
    
data.sort(key=sortFunc)

n=1
h2 = headers[-n:] + headers[:-n]
c2 = colLen[-n:] + colLen[:-n]

with open('PeopleTrainngDate_updated.txt', 'w') as file:
    file.write("  ".join(f"{item:<{c2[i]}}" for i, item in enumerate(h2)))
    file.write("\n")
    file.write("-"*(sum(c2)+(2*(len(c2)-1))))
    file.write("\n")

    for row in data:
        d2 = row[-n:] + row[:-n]
        file.write("  ".join(f"{item:<{c2[i]}}" for i, item in enumerate(d2)))
        file.write("\n")
```

```python
# Exercise 3
data2 = []
with open('PeopleTrainingDateUpdate.csv', 'r') as file:
    csv_reader = csv.reader(file)    
    for row in csv_reader:
        if len(row) == 0:
            continue
        r = []
        r.append(row[0])
        r.append(row[2])
        r.append(row[3])
        m = row[1].split(" ")
        r.append(m[1])
        r.append(m[0])
        r.append(row[4])
        data2.append(r)

with open('PeopleTrainngDate_updated.txt', 'a') as file:
    for row in data2:
        file.write("  ".join(f"{item:<{c2[i]}}" for i, item in enumerate(row)))
        file.write("\n")
```
